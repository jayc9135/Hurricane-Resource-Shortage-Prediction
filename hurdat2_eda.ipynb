```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/hurdat2_formatted.csv')
```

```python
print(len(df))
```

Data with only observation after year 1900

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df_after_1900 = df[df['datetime'].dt.year >= 1900]
```

```python
df_after_1900.head(5)
```

```python
df_storm = df_after_1900[(df_after_1900['datetime'].dt.year >= 2004) & (df_after_1900['storm_name'] == 'CHARLEY')]
df_storm[['storm_name', 'record_identifier', 'latitude', 'longitude', ]]
```

```python
import folium

m = folium.Map(location=[df_storm.iloc[0]['latitude'], df_storm.iloc[0]['longitude']], zoom_start=5)

for i in range(len(df_storm)):
    folium.Marker(
        location=[df_storm.iloc[i]['latitude'], df_storm.iloc[i]['longitude']],
        popup=f"{df_storm.iloc[i]['storm_name']} - {df_storm.iloc[i]['record_identifier']} - {df_storm.iloc[i]['maximum_sustained_wind_knots']} knots",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)

m
```

```python
import random

def random_color():
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))
```

```python
m1 = folium.Map(location=[df_after_1900.iloc[0]['latitude'], df_after_1900.iloc[0]['longitude']], zoom_start=5)

temp_list = []
curr = df_after_1900.iloc[0]['storm_id']
counter = 0
for i in range(len(df_after_1900)):
    if df_after_1900.iloc[i]['storm_id'] == curr:
        temp_list.append((df_after_1900.iloc[i]['latitude'], df_after_1900.iloc[i]['longitude']))
    else:
        counter += 1
        if counter > 10:
            break
        folium.PolyLine(temp_list, color=random_color(), weight=2.5, opacity=1).add_to(m1)
        temp_list = [(df_after_1900.iloc[i]['latitude'], df_after_1900.iloc[i]['longitude'])]
        curr = df_after_1900.iloc[i]['storm_id']
```
